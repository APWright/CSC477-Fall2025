# Example From Lecture

```python
import pandas as pd
df = data.movies()[['Title','Release_Date','Worldwide_Gross','Production_Budget','IMDB_Rating']]
df = df[df['Worldwide_Gross']>0.0]
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
df = df[(df['Release_Date'] > '1945') & (df['Release_Date'] < '2011')]
```

```python
alt.Chart(df).mark_line(color='green').encode(
    x=alt.X('Release_Date:T',axis=alt.Axis(format="%Y")),
    y=alt.Y('Worldwide_Gross').scale(type="log",domain=(1,1e10))
).properties(
        width=400,
        height=200
    )
```

```python
(alt.Chart(df).mark_line(color='green',interpolate="basis").encode(
    x=alt.X('Release_Date:T',axis=alt.Axis(format="%Y")).bin(maxbins=20),
    y=alt.Y('median(Worldwide_Gross)'))
).properties(
        width=400,
        height=200
    )
```

```python
((alt.Chart(df).mark_line(color='green',interpolate="basis").encode(
    x=alt.X('Release_Date:T',axis=alt.Axis(format="%Y")).bin(maxbins=20),
    y=alt.Y('median(Worldwide_Gross)'))
) + (alt.Chart(df).mark_line(color='red',interpolate="basis").encode(
    x=alt.X('Release_Date:T',axis=alt.Axis(format="%Y")).bin(maxbins=20),
    y=alt.Y('median(Production_Budget)')))).properties(
        width=400,
        height=200
    )
```

```python
df['Profits'] = df['Worldwide_Gross'] - df['Production_Budget']
```

```python
((alt.Chart(df).mark_line(interpolate="basis").encode(
    x=alt.X('Release_Date:T',axis=alt.Axis(format="%Y")).bin(maxbins=20),
    y=alt.Y('median(Profits)'))
)).properties(
        width=400,
        height=200
    )
```

```python
alt.Chart(df).mark_circle().encode(
    color=alt.Color('Profits')
            .scale(
                domain=[df['Profits'].min(),0,df['Profits'].max()],
                range=['red','#FFFFFF00','green']),
    size='Production_Budget',
    tooltip='Title:N',
    y='IMDB_Rating',
    x='Release_Date:T'
  ).interactive()
```
